# Basic Signals

```python
#unit impulse
import numpy as np
import matplotlib.pyplot as plt

#Unit Impulse Function
def d(n):
    return np.where(n==0,1,0)

n=np.arange(-10,11)
plt.stem(n,d(n))
plt.title("Unit Impulse")
plt.xlabel("n")
plt.ylabel("d[n]")
plt.grid()
plt.show()
```

```python
#unit step
import numpy as np
import matplotlib.pyplot as plt

#Unit Step Function
def u(n):
    return np.where(n>=0,1,0)

n=np.arange(-10,11)
plt.stem(n,u(n))
plt.title("Unit Step")
plt.xlabel("n")
plt.ylabel("u[n]")
plt.grid()
plt.show()
```

```python
#unit ramp
import numpy as np
import matplotlib.pyplot as plt

#Unit Ramp Function
def r(n):
    return np.where(n>=0,n,0)

n=np.arange(-10,11)
plt.stem(n,r(n))
plt.title("Unit Ramp")
plt.xlabel("n")
plt.ylabel("r[n]")
plt.grid()
plt.show()
```

# Pulses
Pulses can be constructed by shifting scaling and adding basic signals d(n), u(n) and r(n)

For example a rectangular pulse starting at 0 and ending at 2 can be constructed by

pulse = u(n) - u(n-2)

ie. Subtracting a delayed unit step function from a unit step function

```python
#bipolar pulse
import numpy as np
import matplotlib.pyplot as plt

n=np.arange(-5,15)
#Bipolar pulse by shifting, scaling and adding unit step function
bipolar=u(n)-2*u(n-5)+u(n-10)

plt.stem(n,bipolar)
plt.title("Bipolar Pulse")
plt.xlabel("n")
plt.ylabel("y[n]")
plt.grid()
plt.show()
```

```python
#triangular pulse
import numpy as np
import matplotlib.pyplot as plt

def triangle(n):
    return r(n+5)-2*r(n)+r(n-5)

n=np.arange(-10,11)
plt.stem(n,triangle(n))
plt.title("Triangular signal")
plt.xlabel("n")
plt.ylabel("y[n]")
plt.grid()
plt.show()
```

# Periodic Signals
For periodic signals, generate one period of the signal by shift, scale and add of basic functions. Generate a unit impulse train of desired periodicity. Convolve the envelope with the unit impulse train to get the envelope train. 

```python
#Generation of pulse train/periodic waveforms
import numpy as np
import matplotlib.pyplot as plt
n=np.arange(-25,26)

#Generate a Unit impulse train of periodicity 10
train=d(n-20)+d(n-10)+d(n)+d(n+10)+d(n+20)

#Let's generate the pulse that is to be repeated: In this case, a triangular pulse
tri=triangle(n)

#Convolve the two, and remove padding from the output.
y=np.convolve(train,tri,'same')
plt.stem(n,y)
plt.title("Triangular Train")
plt.xlabel("n")
plt.ylabel("y[n]")
plt.grid()
plt.show()
```
